## Requests & Response 데이터 분석
항목을 클릭하면 아래 그림과 같이 서버와 주고 받은 데이터를 확인할 수 있습니다.

##### Headers Tab의 General
웹 서버의 주소 및 프로토콜이 표시됨
- Request URL : 데이터를 요청한 서버의 주소(url)
- Requests Method : 프로토콜 (Get or Post)
  
##### Headers Tab의 Headers

##### Headers Tab의  Query string parameter 혹은 From Data
웹 서버로의 입력 데이터  
상하한가 웹페이지의 경우 조회 버튼 위에 있는 "시장 구분", "구분" 정보가 입력될 것을 예상할 수 있음

Headers Tab의 General 부분을 참고해서 URL을 정의하고 requests 모듈의 post() 메서드를 사용합니다. Headers Tab의 Request headers와 Frome Data 부분은 딕셔너리로 데이터를 정의하구요. 이를 연결한 파이썬 코드는 다음과 같습니다. headers 부분의 일부 코드는 지면 관계상 생략했습니다. 모든 값을 딕셔너리로 정의하면 됩니다. 

```python
import requests

url = "http://marketdata.krx.co.kr/contents/MKD/99/MKD99000001.jspx"

headers = {
    "Host": "marketdata.krx.co.kr",
    "Origin": "http://marketdata.krx.co.kr",
    "Referer": "http://marketdata.krx.co.kr/mdi",
    "User-Agent": "Mozilla/5.0",
    "X-Requested-With": "XMLHttpRequest"
}

data = {
    "market_gubun": "ALL",
    "updn_typ": 4,
    "pagePath": "/contents/MKD/04/0404/04040100/MKD04040100.jsp",
    "code": "aOQJ3lQzxMOYzbHu7vM+XDdOkiWGWFnNn9nGKFpcq66GT+PYPZrr1a8HrBv8XnbVv7gs09yfrS0J+dS1OSzg2b/gjMwrxpsj7F+mt1D3JE5PAUV0mctfceksGgBmZyNUqU1hPlWSwc3J7AYwp0ADpyzPY9ku0F7KsTlHYu7id12fDyQxkvn2gt4pFZufWWNq"
}
resp = requests.post(url, data=data, headers=headers)
print(resp)
print(resp.text)
```

위 코드를 실행하면 일반적으로 크롬 개발자 도구에서 확인했던 Response 데이터를 확인할 수 있습니다. 응답은 'Response 200' (정상) 상태로 출력됐지만 text 속성값이 비어있는 것이 이상합니다. requests할 때 서버로 요청한 data가 무엇인가 잘못되었다고 추측해볼 수 있고, 그 값들을 하나씩 확인해 보겠습니다.
웹페이지에서 시장 구분(전체/KOSPI/KOSDAQ/KONEX)과 구분 항목(상한가/하한가)을 변경해 보면, market_gubun과 updn_typ 이 함께 변경되는 것을 알 수 있습니다. 두 개의 파라미터로 어떤 데이터를 조회할지 결정하는 겁니다. pagePath는 현재 단계에서 정확히 알 수는 없지만 KRX 웹 서버가 내부적으로 사용하는 주소 정도로 추측할 수 있습니다.

## Generate OTP 
알 수없었던 문자열을 얻기 위한 Request URL을 알 수 있으며, Get 프로토콜을 사용하는 것을 알 수 있습니다. 또한 Get 프로토콜은 주소에 데이터를 표현한다고 했었죠? 아래쪽에 위치한 Query String Parameters의 값들이 주소로 표현된것을 알 수 있습니다.

위와 같은 requests를 서버로 전송하면 다음과 같은 문자열 값이 클라이언트로 전송되는 겁니다. 

<img src="image/08.png" style="width=350; height=180">

이를 파이썬 코드로 표현하면 다음과 같습니다. Post와 동일하게 헤더 까지 함께 넘겨 줘야 합니다. (어떤 웹서버는 헤더를 전송하지 않아도 데이터를 전달해 주지만 대부분의 웹서버는 헤더를 함께 전송해야 합니다.)

```python
import requests

url = "http://marketdata.krx.co.kr/contents/COM/GenerateOTP.jspx?bld=MKD%2F04%2F0404%2F04040100%2Fmkd04040100&name=form&_=1588412643904"

headers = {
    "Host": "marketdata.krx.co.kr",    
    "Referer": "http://marketdata.krx.co.kr/mdi",
    "User-Agent": "Mozilla/5.0",
    "X-Requested-With": "XMLHttpRequest"
}

resp = requests.get(url, headers=headers)
print(resp.text)
```

URL에 서버로 전송하는 파라미터가 포함돼 있어서 복잡해 보입니다. 이를 구분해서 코드의 가독성을 높일 수 있습니다. Query String Parameters의 값들을 딕셔너리로 정의하고 params 파라미터로 전달할 수 있습니다. 위와 아래의 코드는 동일한 결과를 출력합니다.  

```python
import requests

url = "http://marketdata.krx.co.kr/contents/COM/GenerateOTP.jspx"

params = {
    "bld": "MKD/04/0404/04040100/mkd04040100",
    "name": "form",
    "_": "1588412643904"
}

headers = {
    "Host": "marketdata.krx.co.kr",    
    "Referer": "http://marketdata.krx.co.kr/mdi",
    "User-Agent": "Mozilla/5.0",
    "X-Requested-With": "XMLHttpRequest"
}

resp = requests.get(url, headers=headers, params=params)
print(resp.text)
```

## 상/하한가 데이터 얻기
두 번의 요청을 합쳐서 하나의 파이썬 코드로 표현하면 다음과 같습니다. Get 프로토콜로 OTP라는 문자열 값을 얻어오고 해당 문자열 값을 사용해서 Post 프로토콜로 상하한가를 조회합니다.

```python
import requests

#-----------------------------------------------------------------------------
# Generate OTP requests
url = "http://marketdata.krx.co.kr/contents/COM/GenerateOTP.jspx"

params = {
    "bld": "MKD/04/0404/04040100/mkd04040100",
    "name": "form",
    "_": "1588412643904"
}

headers = {
    "Host": "marketdata.krx.co.kr",    
    "Referer": "http://marketdata.krx.co.kr/mdi",
    "User-Agent": "Mozilla/5.0",
    "X-Requested-With": "XMLHttpRequest"
}

resp = requests.get(url, headers=headers, params=params)
otp = resp.text

#-----------------------------------------------------------------------------
# 상하한가 requests
url = "http://marketdata.krx.co.kr/contents/MKD/99/MKD99000001.jspx"

headers = {
    "Host": "marketdata.krx.co.kr",
    "Origin": "http://marketdata.krx.co.kr",
    "Referer": "http://marketdata.krx.co.kr/mdi",
    "User-Agent": "Mozilla/5.0",
    "X-Requested-With": "XMLHttpRequest"
}

data = {
    "market_gubun": "ALL",
    "updn_typ": 4,
    "pagePath": "/contents/MKD/04/0404/04040100/MKD04040100.jsp",
    # 중요 : 앞서 generate OTP에서 조회한 otp 값을 두 번째 요청으로 입력합니다. 
    "code": otp
}
resp = requests.post(url, data=data, headers=headers)
print(resp)
print(type(resp.text))
print(resp.text)
```

웹 페이지에서 확인했던 자안, 유진투자증권 등의 데이터가 딕셔너리 형태의 문자열 (json format)로 출력되는 것을 알 수 있습니다. 

response 객체의 json() 메서드는 json을 딕셔너리로 변경합니다. 이 후 반복문으로 모든 데이터의 종목코드, 종목명, 현재가를 출력합니다. 

```python
result = resp.json()
for data in result['DS1']:
    print(data['isu_cd'], data['kor_shrt_isu_nm'], data['isu_cur_pr'])
```
